```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map(df)
```

```python
df=df.drop(['CUST_ID'],axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python

data=scaler.fit_transform(df)
df=pd.DataFrame(data=data, columns=df.columns)
```

```python
df.head()
```

```python
from sklearn.cluster import AgglomerativeClustering 
model=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
clust_labels=model.fit_predict(df)  
df['cluster']=clust_labels

```

```python
df.cluster.value_counts()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="ward"),p=6,truncate_mode='level')
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=df, y='BALANCE', x='PURCHASES', hue= 'cluster')
```

```python
from sklearn.cluster import KMeans  
kmeans=KMeans(n_clusters=4, random_state=0)  
kmeans.fit(df)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(df)
print(kmeans.cluster_centers_)
```

```python
sum_of_squared_distances=[]
K = range(1,10)
for k in K:
    km= KMeans(n_clusters=k)
    km=km.fit(df)
    sum_of_squared_distances.append(km.inertia_)
```

```python

plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt('sum_of_squared_distances')
plt.title('elbow method for optimal k')
plt.show()
```

the optimal k for kmeans based on the elbow method is 4
